```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data_df = pd.read_csv('data.csv',header=None, names=['x1', 'x2', 'y'])
```

```python
data_df.head()
```

```python
def plot_3d_figure(col_a, col_b, col_y):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(col_a, col_b, col_y, c=np.arange(len(col_a)), s=10)
    plt.show()
```

```python
def plot_2d_figure(x, y):
    fig = plt.figure()
    ax = plt.axes()
    ax.scatter(x, y, c=np.arange(len(x)))
    plt.show()
```

```python
plot_2d_figure(data_df.x1, data_df.x2)
```

```python
plot_3d_figure(data_df.x1, data_df.x2, data_df.y)
```

```python
train_dataset, test_dataset = train_test_split(data_df, test_size=0.2, shuffle=True)
```

```python
train_dataset.to_csv('train_dataset.csv', index=False)
```

```python
test_dataset.to_csv('test_dataset.csv', index=False)
```
